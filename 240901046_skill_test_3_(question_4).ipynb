```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset
X = np.array([2010,2011,2012,2013,2014,2015,2016,2017,2018]).reshape(-1,1)
y = np.array([25,28,30,34,37,40,43,45,47])

# Model
model = LinearRegression()
model.fit(X, y)

# Prediction
new_year = np.array([[2019]])
predicted_price = model.predict(new_year)

print("Predicted Price:", predicted_price[0], "Lakhs")

# Graph
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Year")
plt.ylabel("Price (Lakhs)")
plt.title("Year vs House Price")
plt.show()
```
