```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Dataset (Voltage, Current)
X = np.array([
    [4.81, 0.87],
    [5.24, 0.91],
    [4.77, 0.95],
    [4.60, 1.08],
    [4.81, 0.87],
    [12.38, 2.25],
    [12.36, 2.38],
    [11.22, 1.89],
    [12.17, 1.82],
    [12.06, 2.29]
])

# Apply KMeans (2 clusters)
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)

# Get cluster labels
labels = kmeans.labels_

print("Cluster Labels:", labels)

# Plot graph
plt.scatter(X[:,0], X[:,1], c=labels)
plt.xlabel("Voltage")
plt.ylabel("Current")
plt.title("K-Means Clustering")
plt.show()
```
